Machine Learning Model

````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
````

````python
x = iris_data.drop('Species', axis=1)
y = iris_data['Species']
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = RandomForestClassifier(random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
````

````python
print(confusion_matrix(y_test, y_pred))
````

````python
print(classification_report(y_test, y_pred))
````

## Model Performance Analysis

### Overall Accuracy: 100%
The RandomForestClassifier achieved perfect classification on the test set (30 samples).

### What This Means:
* **Zero false positives**: No samples incorrectly labeled as a species they're not
* **Zero false negatives**: No samples missed or misclassified
* **Robust separation**: The model learned clear decision boundaries between all three species

### Model Configuration:
* Algorithm: Random Forest (ensemble of decision trees)
* Train/Test Split: 80/20 (120 training samples, 30 test samples)
* Random State: 42 (ensures reproducibility)
* Features Used: SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm
